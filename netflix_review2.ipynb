```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torchtext.data.utils import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy as np
from torch.nn.utils.rnn import pad_sequence

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï†ÑÏ≤òÎ¶¨
netflix_df = pd.read_csv('netflix_reviews.csv')  # ÎÑ∑ÌîåÎ¶≠Ïä§ Î¶¨Î∑∞ Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ Î∂àÎü¨ÏòµÎãàÎã§.
netflix_df['content'] = netflix_df['content'].fillna('')  # 'content' Ïó¥Ïùò Í≤∞Ï∏°Í∞íÏùÑ Îπà Î¨∏ÏûêÏó¥Î°ú Ï±ÑÏõÅÎãàÎã§.
netflix_df['score'] = netflix_df['score'].fillna('')  # 'score' Ïó¥Ïùò Í≤∞Ï∏°Í∞íÏùÑ Îπà Î¨∏ÏûêÏó¥Î°ú Ï±ÑÏõÅÎãàÎã§.
netflix_df = netflix_df[netflix_df['content'].str.len() > 1]  # Î¶¨Î∑∞Ïùò Í∏∏Ïù¥Í∞Ä 1Î≥¥Îã§ ÌÅ∞ Îç∞Ïù¥ÌÑ∞Îßå ÏÑ†ÌÉùÌï©ÎãàÎã§.

# ÌÜ†ÌÅ∞Ìôî Î∞è Ïñ¥ÌúòÏßë Íµ¨Ï∂ï
tokenizer = get_tokenizer('basic_english')  # Í∏∞Î≥∏ ÏòÅÏñ¥ ÌÜ†ÌÅ¨ÎÇòÏù¥Ï†ÄÎ•º ÏÇ¨Ïö©Ìï©ÎãàÎã§.
def yield_tokens(data_iter):  # Îç∞Ïù¥ÌÑ∞ Î∞òÎ≥µÏûêÏóêÏÑú ÌÜ†ÌÅ∞ÏùÑ ÏÉùÏÑ±ÌïòÎäî Ìï®ÏàòÏûÖÎãàÎã§.
    for text in data_iter:
        yield tokenizer(text)

vocab = build_vocab_from_iterator(yield_tokens(netflix_df['content']), specials=["<unk>"])  # Ïñ¥ÌúòÏßëÏùÑ ÏÉùÏÑ±Ìï©ÎãàÎã§.
vocab.set_default_index(vocab["<unk>"])  # "<unk>" ÌÜ†ÌÅ∞Ïùò Í∏∞Î≥∏ Ïù∏Îç±Ïä§Î•º ÏÑ§Ï†ïÌï©ÎãàÎã§.

# Ï†êÏàò Ïù∏ÏΩîÎî©
label_encoder = LabelEncoder()  # Î†àÏù¥Î∏î Ïù∏ÏΩîÎçîÎ•º ÏÇ¨Ïö©Ìï©ÎãàÎã§.
netflix_df['score'] = label_encoder.fit_transform(netflix_df['score'])  # Ï†êÏàòÎ•º Ïù∏ÏΩîÎî©Ìï©ÎãàÎã§.

# ÌõàÎ†®-ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Î∂ÑÌï†
train_df, test_df = train_test_split(netflix_df, test_size=0.2, random_state=42)  # 80%Îäî ÌõàÎ†® Îç∞Ïù¥ÌÑ∞, 20%Îäî ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î°ú Î∂ÑÌï†Ìï©ÎãàÎã§.

# Ïª§Ïä§ÌÖÄ Îç∞Ïù¥ÌÑ∞ÏÖã
class ReviewDataset(Dataset):  # Îç∞Ïù¥ÌÑ∞ÏÖã ÌÅ¥ÎûòÏä§ Ï†ïÏùò
    def __init__(self, df):
        self.texts = df['content'].tolist()  # ÌÖçÏä§Ìä∏ Î¶¨Ïä§Ìä∏Î•º Í∞ÄÏ†∏ÏòµÎãàÎã§.
        self.labels = df['score'].tolist()  # Î†àÏù¥Î∏î Î¶¨Ïä§Ìä∏Î•º Í∞ÄÏ†∏ÏòµÎãàÎã§.

    def __len__(self):
        return len(self.texts)  # Îç∞Ïù¥ÌÑ∞ÏÖãÏùò Í∏∏Ïù¥Î•º Î∞òÌôòÌï©ÎãàÎã§.

    def __getitem__(self, idx):
        text = self.texts[idx]  # Ïù∏Îç±Ïä§Ïóê Îî∞Îùº ÌÖçÏä§Ìä∏Î•º Í∞ÄÏ†∏ÏòµÎãàÎã§.
        label = self.labels[idx]  # Ïù∏Îç±Ïä§Ïóê Îî∞Îùº Î†àÏù¥Î∏îÏùÑ Í∞ÄÏ†∏ÏòµÎãàÎã§.
        return torch.tensor(vocab(tokenizer(text)), dtype=torch.long), torch.tensor(label, dtype=torch.float)  # ÌÖçÏä§Ìä∏ÏôÄ Î†àÏù¥Î∏îÏùÑ Î∞òÌôòÌï©ÎãàÎã§.

# Ìå®Îî©ÏùÑ ÏúÑÌïú Ïª§Ïä§ÌÖÄ ÏΩúÎ†àÏù¥Ìä∏ Ìï®Ïàò
def collate_fn(batch):  # Î∞∞Ïπò ÎÇ¥ ÌÖçÏä§Ìä∏ÏôÄ Î†àÏù¥Î∏îÏùÑ Ìå®Îî©Ìï©ÎãàÎã§.
    texts, labels = zip(*batch)  # Î∞∞ÏπòÏóêÏÑú ÌÖçÏä§Ìä∏ÏôÄ Î†àÏù¥Î∏îÏùÑ Î∂ÑÎ¶¨Ìï©ÎãàÎã§.
    texts_padded = pad_sequence(texts, batch_first=True)  # ÌÖçÏä§Ìä∏Î•º Ìå®Îî©Ìï©ÎãàÎã§.
    labels_tensor = torch.stack(labels)  # Î†àÏù¥Î∏îÏùÑ ÌÖêÏÑúÎ°ú Î≥ÄÌôòÌï©ÎãàÎã§.
    return texts_padded, labels_tensor  # Ìå®Îî©Îêú ÌÖçÏä§Ìä∏ÏôÄ Î†àÏù¥Î∏îÏùÑ Î∞òÌôòÌï©ÎãàÎã§.

# Îç∞Ïù¥ÌÑ∞Î°úÎçî ÏÉùÏÑ±
train_dataset = ReviewDataset(train_df)  # ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ÏÖã ÏÉùÏÑ±
test_dataset = ReviewDataset(test_df)  # ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ÏÖã ÏÉùÏÑ±
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, collate_fn=collate_fn)  # ÌõàÎ†® Îç∞Ïù¥ÌÑ∞Î°úÎçî ÏÉùÏÑ±
test_loader = DataLoader(test_dataset, batch_size=32, collate_fn=collate_fn)  # ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î°úÎçî ÏÉùÏÑ±

# LSTM Î™®Îç∏ Ï†ïÏùò
class LSTMModel(nn.Module):  # LSTM Î™®Îç∏ ÌÅ¥ÎûòÏä§ Ï†ïÏùò
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        super(LSTMModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)  # ÏûÑÎ≤†Îî© Ï∏µ Ï†ïÏùò
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)  # LSTM Ï∏µ Ï†ïÏùò
        self.fc = nn.Linear(hidden_dim, 1)  # ÏôÑÏ†Ñ Ïó∞Í≤∞ Ï∏µ Ï†ïÏùò

    def forward(self, x):  # ÏàúÏ†ÑÌåå(forward) Ìï®Ïàò Ï†ïÏùò
        embedded = self.embedding(x)  # ÏûÑÎ≤†Îî© Ï∏µ ÌÜµÍ≥º
        lstm_out, _ = self.lstm(embedded)  # LSTM Ï∏µ ÌÜµÍ≥º
        out = self.fc(lstm_out[:, -1, :])  # ÎßàÏßÄÎßâ Ï∂úÎ†•Í∞íÏùÑ ÏôÑÏ†Ñ Ïó∞Í≤∞ Ï∏µÏóê ÌÜµÍ≥º
        return out  # Ï∂úÎ†• Î∞òÌôò

# Î™®Îç∏, ÏÜêÏã§ Ìï®Ïàò, ÏµúÏ†ÅÌôî Í∏∞Î≤ï Ï¥àÍ∏∞Ìôî
vocab_size = len(vocab)  # Ïñ¥ÌúòÏßë ÌÅ¨Í∏∞ ÏÑ§Ï†ï
embedding_dim = 100  # ÏûÑÎ≤†Îî© Ï∞®Ïõê ÏÑ§Ï†ï
hidden_dim = 128  # ÌûàÎì† Ï∞®Ïõê ÏÑ§Ï†ï
model = LSTMModel(vocab_size, embedding_dim, hidden_dim)  # Î™®Îç∏ Ï¥àÍ∏∞Ìôî
criterion = nn.MSELoss()  # ÌèâÍ∑† Ï†úÍ≥± Ïò§Ï∞® ÏÜêÏã§ Ìï®Ïàò ÏÇ¨Ïö©
optimizer = optim.Adam(model.parameters(), lr=0.001)  # Adam ÏµúÏ†ÅÌôî Í∏∞Î≤ï ÏÇ¨Ïö©

# ÌõàÎ†® Î£®ÌîÑ
num_epochs = 5  # ÏóêÌè¨ÌÅ¨ Ïàò ÏÑ§Ï†ï
for epoch in range(num_epochs):
    model.train()  # Î™®Îç∏ÏùÑ ÌõàÎ†® Î™®ÎìúÎ°ú ÏÑ§Ï†ï
    for texts, labels in train_loader:  # ÌõàÎ†® Îç∞Ïù¥ÌÑ∞Î°ú Î∞òÎ≥µ
        optimizer.zero_grad()  # Í∏∞Ïö∏Í∏∞ Ï¥àÍ∏∞Ìôî
        outputs = model(texts)  # Î™®Îç∏ Ï∂úÎ†• Í≥ÑÏÇ∞
        loss = criterion(outputs.squeeze(), labels)  # ÏÜêÏã§ Í≥ÑÏÇ∞
        loss.backward()  # Ïó≠Ï†ÑÌåå
        optimizer.step()  # Í∞ÄÏ§ëÏπò ÏóÖÎç∞Ïù¥Ìä∏
    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {loss.item()}')  # ÏóêÌè¨ÌÅ¨ÏôÄ ÏÜêÏã§ Ï∂úÎ†•

# 11. ÌèâÍ∞Ä
model.eval()
with torch.no_grad():
    total_loss = 0
    for texts, labels in test_loader:
        outputs = model(texts)
        loss = criterion(outputs.squeeze(), labels)
        total_loss += loss.item()
    avg_loss = total_loss / len(test_loader)
    print(f'Average Test Loss: {avg_loss}')

# ÏÉàÎ°úÏö¥ Î¶¨Î∑∞Ïóê ÎåÄÌïú ÏòàÏ∏°
new_review = "This app is great but has some bugs."  # ÏÉàÎ°úÏö¥ Î¶¨Î∑∞

# Î¶¨Î∑∞Î•º ÌÜ†ÌÅ∞ÌôîÌïòÍ≥† Ïù∏Îç±Ïä§Î°ú Î≥ÄÌôò
tokenized_review = vocab(tokenizer(new_review))  # ÌÜ†ÌÅ∞Ìôî Î∞è Ïñ¥ÌúòÏßë Ïù∏Îç±Ïä§ Î≥ÄÌôò
input_tensor = torch.tensor(tokenized_review, dtype=torch.long).unsqueeze(0)  # Î∞∞Ïπò Ï∞®Ïõê Ï∂îÍ∞Ä

# ÏòàÏ∏° Ï†êÏàò Í≥ÑÏÇ∞
model.eval()  # Î™®Îç∏ÏùÑ ÌèâÍ∞Ä Î™®ÎìúÎ°ú ÏÑ§Ï†ï
with torch.no_grad():  # Í∏∞Ïö∏Í∏∞ Í≥ÑÏÇ∞ ÎπÑÌôúÏÑ±Ìôî
    predicted_score = model(input_tensor)  # Î™®Îç∏ÏùÑ ÌÜµÌï¥ ÏòàÏ∏°
    predicted_score = predicted_score.item()  # ÌÖêÏÑúÎ•º Ïä§ÏπºÎùº Í∞íÏúºÎ°ú Î≥ÄÌôò

print(f'Predicted Score: {predicted_score}')  # ÏòàÏ∏° Ï†êÏàò Ï∂úÎ†•
```

```python
import pandas as pd
from textblob import TextBlob

from IPython.display import display
# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df = pd.read_csv('netflix_reviews.csv')

# Í∞êÏÑ± Î∂ÑÏÑù Ìï®Ïàò Ï†ïÏùò
def get_sentiment(text):
    if isinstance(text, str):  # Î¨∏ÏûêÏó¥Ïù∏ Í≤ΩÏö∞ÏóêÎßå Î∂ÑÏÑù
        return TextBlob(text).sentiment.polarity
    return 0  # Î¨∏ÏûêÏó¥Ïù¥ ÏïÑÎãê Í≤ΩÏö∞ Í∏∞Î≥∏Í∞í ÏÑ§Ï†ï

# NaN Î∞è ÎπÑÎ¨∏ÏûêÏó¥ Ï≤òÎ¶¨
df['content'] = df['content'].fillna('')  # NaNÏùÑ Îπà Î¨∏ÏûêÏó¥Î°ú ÎåÄÏ≤¥

# Í∞êÏÑ± Î∂ÑÏÑù Ï†ÅÏö©
df['sentiment'] = df['content'].apply(get_sentiment)

# Í∞êÏÑ± ÎùºÎ≤®ÎßÅ
df['sentiment_label'] = df['sentiment'].apply(lambda x: 'positive' if x > 0.1 else ('negative' if x < -0.1 else 'neutral'))


# ÏõêÌïòÎäî Ïó¥Îßå ÏÑ†ÌÉù
result_df = df[['content', 'score', 'sentiment_label']].copy()  # copy() Î©îÏÑúÎìú Ìò∏Ï∂ú
result_df.rename(columns={'content': 'content-c'}, inplace=True)
df.to_string(index=False)
# Jupyter NotebookÏóêÏÑú HTML ÌòïÏãùÏúºÎ°ú Ï∂úÎ†•
```
